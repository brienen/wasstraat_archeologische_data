```python
path = "/input/sdschijf"
lst_ext = ['.pdf', '.doc', '.docx']
excel_bestand = './Alle Bestanden Archeologisch Depot.xlsx'
```

```python
df = pd.DataFrame(columns = ['Bestand', 'Extensie', 'Dir', 'LaatsteDir'])

for root,d_names,f_names in os.walk(path, topdown=True, onerror=None, followlinks=False):
    for file in f_names:
        file_name, file_extension = os.path.splitext(file)
        lastpath = os.path.basename(os.path.normpath(root))
        if file_extension.lower() in lst_ext:
            df_tmp = pd.DataFrame.from_dict({'Bestand' : [file], 'Extensie' : [file_extension], 'Dir' : [root], 'LaatsteDir': [lastpath]})
            df = pd.concat([df, df_tmp])

df
```

```python
df.to_excel(excel_bestand)  
```
